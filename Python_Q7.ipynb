```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import math
```

```python
data = pd.read_csv("Data/meteorite_data (1).csv")
```

```python
data.head()
```

```python
data['year'] = pd.to_datetime(data['year'], errors='coerce')
```

```python
list_point_coordinates = []
for list_str in data['point coordinates']:
    temp_list = []
    list_str_ele = list_str.strip("[]")
    list_str_ele = list_str_ele.split(", ")
    for ele in list_str_ele:
        temp_list.append(int(float(ele)))
    list_point_coordinates.append(temp_list)
```

```python
 data['point coordinates'] = list_point_coordinates
```

## Get all the Earth meteorites that fell before the year 2000

```python
earth_meteorites_before_2000 = data[(data['reclat'].notnull()) & (data['year'].dt.year < 2000)]
earth_meteorites_before_2000
```

## Get all the earth meteorites co-ordinates who fell before the year 1970

```python
earth_meteorites_coords_before_1970 = data[data['year'].dt.year < 1970]
earth_meteorites_coords_before_1970
```

## Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg


```python
earth_meteorites_mass_gt_10000kg = data[data['mass'] > 10000]
earth_meteorites_mass_gt_10000kg
```

##  Plotting the number of Earth meteorites per year before 2000

```python
meteorites_per_year = earth_meteorites_before_2000.groupby(earth_meteorites_before_2000['year'].dt.year).size()
meteorites_per_year.plot(kind='bar', figsize=(10, 6))
plt.title('Number of Earth Meteorites per Year (before 2000)')
plt.xlabel('Year')
plt.ylabel('Number of Meteorites')
plt.show()
```

## Plotting the coordinates of Earth meteorites that fell before 1970

```python
plt.scatter(earth_meteorites_coords_before_1970['reclong'], earth_meteorites_coords_before_1970['reclat'])
plt.title('Coordinates of Earth Meteorites that Fell before 1970')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

##  Plotting the mass distribution of Earth meteorites (>10000 kg) before 2000

```python
plt.hist(earth_meteorites_mass_gt_10000kg['mass'] / 1000, bins=20)
plt.title('Mass Distribution of Earth Meteorites (>10000 kg) before 2000')
plt.xlabel('Mass (kg)')
plt.ylabel('Frequency')
plt.show()
```
