## Q1

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def insertNode(head, data):
    newNode = Node(data)
    if head is None:
        head = newNode
    else:
        curr = head
        while curr.next:
            curr = curr.next
        curr.next = newNode
    return head

def createNewList(list1, list2):
    newHead = None
    curr1 = list1
    curr2 = list2

    while curr1 and curr2:
        if curr1.data >= curr2.data:
            newHead = insertNode(newHead, curr1.data)
        else:
            newHead = insertNode(newHead, curr2.data)
        curr1 = curr1.next
        curr2 = curr2.next

    while curr1:
        newHead = insertNode(newHead, curr1.data)
        curr1 = curr1.next

    while curr2:
        newHead = insertNode(newHead, curr2.data)
        curr2 = curr2.next

    return newHead

def printList(head):
    curr = head
    while curr:
        print(curr.data, end="->")
        curr = curr.next
    print("None")

# Example 1
list1 = None
list1 = insertNode(list1, 5)
list1 = insertNode(list1, 2)
list1 = insertNode(list1, 3)
list1 = insertNode(list1, 8)

list2 = None
list2 = insertNode(list2, 1)
list2 = insertNode(list2, 7)
list2 = insertNode(list2, 4)
list2 = insertNode(list2, 5)

newList = createNewList(list1, list2)
print("New list: ", end="")
printList(newList)

# Example 2
list1 = None
list1 = insertNode(list1, 2)
list1 = insertNode(list1, 8)
list1 = insertNode(list1, 9)
list1 = insertNode(list1, 3)

list2 = None
list2 = insertNode(list2, 5)
list2 = insertNode(list2, 3)
list2 = insertNode(list2, 6)
list2 = insertNode(list2, 4)

newList = createNewList(list1, list2)
print("New list: ", end="")
printList(newList)
```

### Q2

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def removeDuplicates(head):
    curr = head

    while curr and curr.next:
        if curr.data == curr.next.data:
            curr.next = curr.next.next
        else:
            curr = curr.next

    return head

def printList(head):
    curr = head
    while curr:
        print(curr.data, end="->")
        curr = curr.next
    print("None")

# Example
head = None
head = Node(10)
head.next = Node(12)
head.next.next = Node(12)
head.next.next.next = Node(25)
head.next.next.next.next = Node(25)
head.next.next.next.next.next = Node(25)
head.next.next.next.next.next.next = Node(34)

print("Original list: ", end="")
printList(head)

head = removeDuplicates(head)

print("List after removing duplicates: ", end="")
printList(head)
```

### Q3

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverseKNodes(head, k):
    if head is None or k == 1:
        return head

    dummy = Node(0)
    dummy.next = head
    prev = dummy
    curr = head
    count = 0

    while curr:
        count += 1
        if count % k == 0:
            prev = reverseGroup(prev, curr.next)
            curr = prev.next
        else:
            curr = curr.next

    return dummy.next

def reverseGroup(prev, next_node):
    last = prev.next
    curr = last.next

    while curr != next_node:
        last.next = curr.next
        curr.next = prev.next
        prev.next = curr
        curr = last.next

    return last

def printList(head):
    curr = head
    while curr:
        print(curr.data, end="->")
        curr = curr.next
    print("None")

# Example
head = None
head = Node(1)
head.next = Node(2)
head.next.next = Node(2)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)

k = 4

print("Original list: ", end="")
printList(head)

head = reverseKNodes(head, k)

print("List after reversing every", k, "nodes: ", end="")
printList(head)
```

### Q4

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverseAlternateKNodes(head, k):
    if head is None or k <= 1:
        return head

    curr = head
    prev = None
    counter = 0

    # Reverse first k nodes
    while curr and counter < k:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node
        counter += 1

    # Skip next k nodes
    while curr and counter < 2 * k:
        curr = curr.next
        counter += 1

    # Recursive call for remaining linked list
    if curr:
        head.next = reverseAlternateKNodes(curr, k)

    return prev

def printList(head):
    curr = head
    while curr:
        print(curr.data, end="->")
        curr = curr.next
    print("None")

# Example
head = None
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)
head.next.next.next.next.next.next.next.next = Node(9)

k = 3

print("Original list: ", end="")
printList(head)

head = reverseAlternateKNodes(head, k)

print("List after reversing every alternate", k, "nodes: ", end="")
printList(head)
```

### Q5

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def deleteLastOccurrence(head, key):
    if head is None:
        return head

    prev = None
    last = None
    curr = head

    while curr:
        if curr.data == key:
            last = prev
        prev = curr
        curr = curr.next

    if last:
        last.next = last.next.next
    else:
        head = head.next

    return head

def printList(head):
    curr = head
    while curr:
        print(curr.data, end="->")
        curr = curr.next
    print("None")

# Example
head = None
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(5)
head.next.next.next.next = Node(2)
head.next.next.next.next.next = Node(10)

key = 2

print("Original list: ", end="")
printList(head)

head = deleteLastOccurrence(head, key)

print("List after deleting last occurrence of", key, ": ", end="")
printList(head)
```

### Q6

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def mergeSortedLists(headA, headB):
    if headA is None:
        return headB
    if headB is None:
        return headA

    dummy = Node(0)
    tail = dummy

    while headA and headB:
        if headA.data <= headB.data:
            tail.next = headA
            headA = headA.next
        else:
            tail.next = headB
            headB = headB.next
        tail = tail.next

    if headA:
        tail.next = headA
    if headB:
        tail.next = headB

    return dummy.next

def printList(head):
    curr = head
    while curr:
        print(curr.data, end="->")
        curr = curr.next
    print("None")

# Example 1
a = None
a = Node(5)
a.next = Node(10)
a.next.next = Node(15)

b = None
b = Node(2)
b.next = Node(3)
b.next.next = Node(20)

print("List A: ", end="")
printList(a)
print("List B: ", end="")
printList(b)

merged = mergeSortedLists(a, b)

print("Merged list: ", end="")
printList(merged)

# Example 2
a = None
a = Node(1)
a.next = Node(1)

b = None
b = Node(2)
b.next = Node(4)

print("List A: ", end="")
printList(a)
print("List B: ", end="")
printList(b)

merged = mergeSortedLists(a, b)

print("Merged list: ", end="")
printList(merged)
```

### Q7

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def reverseDoublyLinkedList(head):
    if head is None or head.next is None:
        return head

    curr = head
    temp = None

    while curr:
        temp = curr.prev
        curr.prev = curr.next
        curr.next = temp
        curr = curr.prev

    head = temp.prev
    return head

def printList(head):
    curr = head
    while curr:
        print(curr.data, end=" ")
        curr = curr.next
    print()

# Example
head = None
head = Node(10)
head.next = Node(8)
head.next.prev = head
head.next.next = Node(4)
head.next.next.prev = head.next
head.next.next.next = Node(2)
head.next.next.next.prev = head.next.next

print("Original list:", end=" ")
printList(head)

head = reverseDoublyLinkedList(head)

print("Reversed list:", end=" ")
printList(head)
```

### Q8

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def deleteNodeAtPosition(head, position):
    if head is None:
        return head

    if position == 1:
        if head.next is not None:
            head.next.prev = None
        head = head.next
        return head

    curr = head
    count = 1

    while curr and count < position:
        curr = curr.next
        count += 1

    if curr is None:
        return head

    if curr.next is not None:
        curr.next.prev = curr.prev

    curr.prev.next = curr.next

    return head

def printList(head):
    curr = head
    while curr:
        print(curr.data, end=" ")
        curr = curr.next
    print()

# Example
head = None
head = Node(1)
head.next = Node(3)
head.next.prev = head
head.next.next = Node(4)
head.next.next.prev = head.next

print("Original list:", end=" ")
printList(head)

position = 3

head = deleteNodeAtPosition(head, position)

print("List after deleting node at position", position, ":", end=" ")
printList(head)
```
