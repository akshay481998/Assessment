```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def download_data_from_link(link):
    response = requests.get(link)
    df = pd.read_csv(response.url)
    return df
```

```python
link = "https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD"
df = download_data_from_link(link)
```

## Get all the cars and their types that do not qualify for clean alternative fuel vehicle.

```python
non_clean_cars = df[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] == 'Not eligible due to low battery range'][['Make', 'Model']]
print("Cars and Types that do not qualify for clean alternative fuel vehicle:")
non_clean_cars
```

## Get all TESLA cars with the model year, and model type made in Bothell City.

```python
tesla_cars_bothell = df[(df['Make'] == "TESLA") & (df["City"] == "Bothell")][['Model Year','Model']]
tesla_cars_bothell
```

## Get all the cars that have an electric range of more than 100, and were made after 2015.

```python
electric_cars_range_100 = df[(df['Electric Range']>100) & (df["Model Year"]>2015)]
electric_cars_range_100
```

## Draw plots to show the distribution between city and electric vehicle type

```python
city_ev_type_distribution = df.groupby(['City', 'Electric Vehicle Type']).size().unstack().fillna(0)
city_ev_type_distribution.plot(kind='bar', stacked=True)
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Distribution of Electric Vehicle Types by City')
plt.show()
```
