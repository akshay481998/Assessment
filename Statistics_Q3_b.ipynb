```python
import pandas as pd
import statistics
import matplotlib.pyplot as plt

# let's import the data
df = pd.read_csv("Data/data.csv")

# To calculate the confidence interval, first we have to calculate the differences between before and after blood pressure.
differences = df[' Blood Pressure Before (mmHg)'] - df[' Blood Pressure After (mmHg)']

# Calculate mean of the differences.
difference_mean = statistics.mean(differences)
print("Mean of the differences: ",difference_mean)
print()

# Calculate standard deviation of differences
differences_std_dev = statistics.stdev(differences)
print("Standard deviation of differences: ",differences_std_dev)
print()

# Calculate standard error of differences
SE = differences_std_dev / (len(differences) ** 0.5)
print("Standard Error: ",SE)
print()

# Calculate margin of error
ME = 1.96 * SE
print("Margine of Error: ",ME)
print()

# Calculate confidence interval
lower_bound = difference_mean - ME
upper_bound = difference_mean + ME
print("lower bound: ",lower_bound)
print()
print("upper bound: ",upper_bound)
print()

# Plot the mean and confidence interval
plt.figure()
plt.axhline(difference_mean, color="blue", label="Mean")
plt.axhline(lower_bound, color="red", linestyle='--', label='Lower Bound (5% CI)')
plt.axhline(upper_bound, color="red", linestyle='--', label='Upper Bound (5% CI)')
plt.legend()
plt.xlabel("Difference between before and after blood pressure")
plt.title('Mean and 5% Confidence Interval of Blood Pressure')
plt.show()
```
